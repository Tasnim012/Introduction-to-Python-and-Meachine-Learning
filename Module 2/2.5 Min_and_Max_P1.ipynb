```python
A = int(numbers[0])
B = int(numbers[1])
C = int(numbers[2])

Min = A
Max = A
if B < Min:
    Min = B
if C < Min:
    Min = C  
    
if B > Max:
    Max = B
if C > Max:
    Max = C  

print(Min, Max)  
```
