```python
list = []
type(list)
```

```python
stack = []

#stack a push
#Last is First out (LIFO)
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)

print(stack)
```

```python
#getting the top element
print(f"Top Element: {stack[-1]}")

#remove the top element
stack.pop()

#new top element
print(f"New Top Element: {stack[-1]}")
```

List as Queue

```python
queue = []

#First in First out(FIFO)
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

print(queue)
```

```python
#remove
queue.pop(0)
print(queue)
```

```python
#Access front element
print(f"Front element {queue[0]}")
```
