```python
string = "Welcome to Phitron ML Course"

#length
print(len(string))

#Find any word in string 
print("Phitron" in string)
```

```python
string = "Welcome to Phitron ML Course"

processed_string = string.lower()
print(processed_string)

#Find any word in string
print("phitron" in processed_string)
```

```python
string = "Welcome to Phitron ML Course"

processed_string = string.upper()
print(processed_string)

#Find any word in string
print("PHITRON" in processed_string)
```

```python
string = "Welcome to Phitron ML Course, Phitron ML, Hello ML"

processed_string = string.capitalize()
print(processed_string)

#Find any word in string
print("phitron" in processed_string)
```

```python
#shurur kon index theke substring ta shuru hoice
shuru_index = processed_string.find("phitron")
print(shuru_index)

#shesher dike kon index theke substring ta shuru hoice
shesh_index = processed_string.rfind("phitron")
print(shesh_index)

#Substring koibar ache
count = processed_string.count("phitron")
print(count)

#replace a substring
new_string = string.replace("ML", "AI/ML")
print(new_string)
```
