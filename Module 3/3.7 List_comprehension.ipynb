```python
#List of even numbers
even = []
#naive
for i in range(20):
    if i % 2 == 0 and i != 0:
        even.append(i)
print(even)
```

List comprehension

```python
random_list = [x for x in range(50) if x%2==0 and x != 0]
print(random_list)
```

```python
random_list = [x for x in range(50) if x%2!=0 and x != 0]
print(random_list)
```
