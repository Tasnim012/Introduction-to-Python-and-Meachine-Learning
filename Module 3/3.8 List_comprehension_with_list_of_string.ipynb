```python
fruits = ['apple', 'orange', 'mango']
capitalize_fruits = []

for fruit in fruits:
    capitalize_fruits.append(fruit.capitalize())
print(capitalize_fruits)
```

```python
fruits = ['apple', 'orange', 'mango']
upper_fruits = []

for fruit in fruits:
    upper_fruits.append(fruit.upper())
print(upper_fruits)
```

Using list comprehension

```python
fruits = ['APPLE', 'ORANGE', 'MANGO']
lower_fruits = [fruit.lower() for fruit in fruits]

print(lower_fruits)
```
